```python
# matriz de coeficientes C
C = data.iloc[:19,:19]
C=C.apply(pd.to_numeric, errors='coerce')
C
```

```python
# total producciones
total_p = data.iloc[:19,19:20]
total_p = total_p.apply(pd.to_numeric, errors='coerce')
total_p
```

```python
total_p = total_p.squeeze()
```

```python
# vector de demanda final d
vector_d = data.iloc[:19,20:21]
vector_d
```

## <font color=#d4b1e6  > **Visualización Matriz** </font>

```python
print("Forma de C:", C.shape)
print("Forma de total_p:", total_p.shape)
```

```python
C_final = C.div(total_p, axis=0)
C_final = C_final.fillna(0)
C_final
```

```python
I = np.eye(C_final.shape[0])  
I
```

```python
A = I - C_final
A = np.nan_to_num(A)
```

## <font color=#d4b1e6  > **Gauss-Jordan** </font>

```python
def gauss_jordan(A, b):
    matriz = np.concatenate((A, b), axis=1)
    filas = matriz.shape[0]
    
    for i in range(filas):
        for j in range(i+1, filas):
            factor = -matriz[j, i] / matriz[i, i]
            matriz[j, :] = matriz[i, :] * factor + matriz[j, :]
            
    for i in range(filas-1, -1, -1):
        for j in range(i-1, -1, -1):
            factor = -matriz[j, i] / matriz[i, i]
            matriz[j, :] = matriz[i, :] * factor + matriz[j, :]
        
        matriz[i, :] = matriz[i, :] / matriz[i, i]
    
    x = matriz[:, -1]  # Última columna tiene las soluciones
    return x
```

```python
tiempo_inicial = time.time()
x_gauss = gauss_jordan(A, vector_d.values)
tiempo_final = time.time()
tiempo_gauss = tiempo_final - tiempo_inicial

print("Solución con método de Gauss:", x_gauss)
print(f"Tiempo del método de Gauss: {tiempo_gauss:.6f} segundos")
```

## <font color=#d4b1e6  > **Comprobación con función 'inv'** </font>

```python
tiempo_inicial = time.time()
A_inv = np.linalg.inv(A)  
x_inv = np.dot(A_inv, vector_d.values)
tiempo_final = time.time()
print("Solución usando la inversa:", x_inv)
print(f"Tiempo de inversa: {tiempo_gauss:.6f} segundos")
```

## <font color=#d4b1e6  > **Método de Gauss-Seidel** </font>

```python
def gauss_seidel(A, b, tol=1**-20, max_iter=10000):
    n = len(b)
    x = np.zeros(n)  

    for k in range(max_iter):
        x_prev = np.copy(x)
        for i in range(n):
            sum_Ax_antes = sum(A[i][j] * x[j] for j in range(i))
            sum_Ax_desp = sum(A[i][j] * x_prev[j] for j in range(i + 1, n))
            x[i] = (b[i] - sum_Ax_antes - sum_Ax_desp) / A[i][i]
        
        norm_diff = np.linalg.norm(x - x_prev, ord=np.inf)
        norm_x = np.linalg.norm(x, ord=np.inf)
        
        if norm_diff / norm_x < tol:
            for idx, val in enumerate(x, 1):
                print(f'x{idx} = {val:.6f}')
            return x, k + 1

    print('No se puede calcular la solución')
    return None, max_iter
```

```python
tiempo_inicial_gs = time.time()
x_g_s = gauss_seidel(A, vector_d.values)
tiempo_final_gs = time.time()
tiempo_gs = tiempo_final_gs - tiempo_inicial_gs

print("Solución con método de Gauss-Seidel:", x_g_s)
print(f"Tiempo del método de Gauss-Seidel: {tiempo_gs:.6f} segundos")
```
